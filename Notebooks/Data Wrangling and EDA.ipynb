```python
data.corr()
```

```python
sns.pairplot(data)
plt.show()
```

a given cell value represents the correlation that exists between two variables on the diagonal, you can see a bunch of histograms. This is because pairplotting the variables with themselves would be pointless, so the pairplot() method instead makes histograms to show the distributions of those variables' values. This allows us to quickly see the shape of each variable's values.

```python
# Make a heatmap of the data 
plt.figure(figsize=(15, 10))

# Create an annotated heatmap of correlations
sns.heatmap(data.corr(), annot=True)

plt.title("Correlation Matrix")

# Show the plot
plt.show()
```
